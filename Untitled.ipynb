```python
import os
import pandas as pd
import json
from utils import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
video_data = [f for f in os.listdir('data') if 'video' in f]
video_data
```

```python
len(video_data)
```

```python
sub_df = []
for f in video_data:
    with open(f'data/{f}', 'r') as js:
        data = json.load(js)

    sub_df.append(pd.json_normalize(data, record_path = ['items']))

df = pd.concat(sub_df)
```

```python
df.shape
```

```python
df.head()
```

```python
cols = [col for col in df.columns if 'thumbnail' not in col]
df2 = df[cols].copy()
```

```python
int_cols = ['statistics.viewCount', 'statistics.likeCount', 'statistics.favoriteCount', 'statistics.commentCount']
for col in int_cols:
    df2[col].fillna(0, inplace = True)
    df2[col] = df2[col].astype(int)
df2
```

```python
sns.scatterplot(x = 'statistics.viewCount', y = 'statistics.likeCount', data = df2)
plt.show()
```
