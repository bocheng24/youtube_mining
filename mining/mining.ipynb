```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from datasets import Dataset
```

```python
dataset = Dataset()
```

```python
dataset.video.head()
```

```python
dataset.channel
```

```python
plt.figure(figsize = (12, 6))
sns.scatterplot(dataset.channel, x = 'videoCount', y = 'subscriberCount')
plt.xticks(np.linspace(dataset.channel['videoCount'].min(), dataset.channel['videoCount'].max(), 25), rotation = 45)
plt.yticks(np.linspace(dataset.channel['subscriberCount'].min(), dataset.channel['subscriberCount'].max(), 25))
plt.show()
```

```python
dataset.channel['subscriberCount'].median()
```

```python
dataset.channel['videoCount'].median()
```

```python
dataset.video['type'] = dataset.video['duration'].apply(lambda x : 'shorts' if x <= 60 else 'long')
```

```python
dataset.video.groupby(['type'])['type'].count()
```

```python
shorts = dataset.video[dataset.video['type'] == 'shorts']
```

```python
plt.figure(figsize = (12, 8))
sns.countplot(shorts, x = 'duration')
plt.xticks(rotation = 45)
plt.show()
```

```python
shorts.groupby('duration')['viewCount'].mean()
```
