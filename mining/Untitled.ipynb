```python
cydf = YTBDF(json_data, 'channel')
ch_df = cydf.get_all_json()
```

```python
ch_df['country'].unique()
```

```python
ch_df['country'] = ch_df['country'].fillna('unknown')

plt.figure(figsize = (12, 8))
sns.countplot(ch_df, x = 'country')
plt.xticks(rotation = 45)
plt.show()
```

```python
country = ch_df.groupby('country').agg({'viewCount': ['mean', 'median', 'count']})

country.columns = ['_'.join(col) for col in country.columns]
country = country.sort_values(by = ['viewCount_count'], ascending = False)
country
```

```python
plt.figure(figsize = (12, 8))
sns.barplot(country, x = country.index, y = 'viewCount_count')
plt.xticks(rotation = 45)
plt.show()
```

```python
us = ch_df[ch_df['country'] == 'US'].copy()
plt.figure(figsize = (12, 8))
sns.scatterplot(us, x = 'videoCount', y = 'viewCount')

plt.show()
```

```python
us['perVideoCount'] = us['viewCount'] / us['videoCount']
us.sort_values(by = 'perVideoCount', ascending = False)[['title', 'viewCount', 'videoCount', 'subscriberCount', 'perVideoCount']]
```
